# Exploracion inicial - OIMET_OFERTAS

Carga del dataset `OIMET_OFERTAS.csv` y revision inicial de estructura, tipos de datos y columnas con valores vacios.

```python
from pathlib import Path

import pandas as pd

dataset_path = Path("dataset/OIMET_OFERTAS.csv")
if not dataset_path.exists():
    dataset_path = Path("project/dataset/OIMET_OFERTAS.csv")

df_oimet = pd.read_csv(dataset_path, sep=";", encoding="utf-8-sig")

print(f"Dataset cargado desde: {dataset_path}")
print(f"Filas: {df_oimet.shape[0]:,}")
print(f"Columnas: {df_oimet.shape[1]:,}")
df_oimet.head()
```

## Columnas vacias

Conteo de registros vacios por columna y cantidad de filas que tienen al menos una columna vacia.

```python
nulos_por_columna = df_oimet.isna().sum()

resumen_nulos = (
    nulos_por_columna
    .to_frame(name="registros_vacios")
    .assign(porcentaje=lambda tabla: (tabla["registros_vacios"] / len(df_oimet) * 100).round(2))
    .sort_values("registros_vacios", ascending=False)
)

filas_con_columnas_vacias = df_oimet.isna().any(axis=1).sum()

print(f"Filas con al menos una columna vacia: {filas_con_columnas_vacias:,} de {len(df_oimet):,}")
display(resumen_nulos)
```
