```python
import geopandas as gpd

berlin = gpd.read_file("https://raw.githubusercontent.com/ljwolf/geopython/refs/heads/master/data/berlin-neighbourhoods.geojson")
```

```python
berlin.area
```

Encontrando el centroide de cada poligono

```python
berlin["puntos"] = berlin.geometry.centroid
berlin["puntos"]
```

Graficando

```python
berlin_punto = berlin.set_geometry("puntos")
berlin_punto.plot()
```

```python
berlin.plot()
```
